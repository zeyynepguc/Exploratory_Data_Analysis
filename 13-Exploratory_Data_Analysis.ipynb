```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('WineQT.csv')
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="fixed acidity", y="density", data=data)
plt.xlabel("Fixed Acidity")
plt.ylabel("Density")
plt.title("Wines of Fixed Acidity vs Density")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='alcohol', kde=True, color='darkred', bins=25)
plt.xlabel("Alcohol Content(%)")
plt.ylabel("Frequency")
plt.title("Distribution of Alcohol Content")
plt.show()
```

```python
data.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Wine Variables")
plt.show()
```

```python
data.groupby("quality").mean()
```

```python
data["quality"].value_counts().plot(kind="bar")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(data["pH"])
plt.xlabel("pH")
plt.ylabel("Count")
plt.title("Distribution of pH")
plt.show()
```

```python
sns.catplot(x="quality", y="alcohol",hue="density", data=data)
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=data, x="quality", y="alcohol")
plt.xlabel("Wine Quality")
plt.ylabel("Alcohol")
plt.title("Alcohol Content by Wine Quality")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="fixed acidity", y="density", hue="quality", data=data)
plt.xlabel("Fixed Acidity")
plt.ylabel("Density")
plt.title("Fixed Acidity vs Density Colored by Quality", alpha=0.7)
plt.legend(title='Quality')
plt.show()
```

```python
sns.barplot(x="quality", y="volatile acidity", data=data)
plt.show()
```
